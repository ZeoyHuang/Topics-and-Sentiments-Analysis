```python
from snownlp import SnowNLP
import pandas as pd
data=pd.read_csv('分时数据\上海疫情（结构化地址）.csv',encoding = 'utf-8')
data.head()
```

```python
data1=data[['微博正文','发布时间']]
data1.head()
```

```python
from snownlp import SnowNLP
data['emotion']=data1['emotion'] = data1['微博正文'].apply(lambda x: SnowNLP(x).sentiments)
```

```python
import matplotlib.pyplot as plt
import numpy as np
 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
 
bins = np.arange(0, 1.1, 0.01)
plt.hist(data1['emotion'], bins, color = '#4F94CD', alpha=0.9)
plt.xlim(0, 1)
plt.xlabel('情感分')
plt.ylabel('数量')
plt.title('情感分直方图')
plt.show()
```

```python
#jieba提取关键词
import jieba
from jieba import analyse
 
stop_words = stopwords = [line.strip() for line in open('stopwords-master\stopwords.txt', 'r',encoding='utf-8').readlines()]
text = ''
for i in data1['微博正文']:
    if i not in stop_words:
        text+=i
    
key_words = jieba.analyse.set_stop_words('stopwords-master\stopwords.txt')    #设置停用词
key_words = jieba.analyse.extract_tags(sentence=text, topK=20, withWeight=True, allowPOS=())

key_words
```

```python
key_words = jieba.analyse.extract_tags(sentence=text, topK=10, withWeight=True, allowPOS=())
key_words
```

```python
pos, neg = 0, 0
for i in data1['emotion']:
    if i >= 0.5:
        pos += 1
    else:
        neg += 1
print('积极评论数目为:', pos, '\n消极评论数目为：', neg)
```

```python

import matplotlib.pyplot as plt
 
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
 
pie_labels = 'positive', 'negative'
plt.pie([pos, neg], labels=pie_labels, autopct='%1.2f%%', shadow=True)
 
plt.show()
```
