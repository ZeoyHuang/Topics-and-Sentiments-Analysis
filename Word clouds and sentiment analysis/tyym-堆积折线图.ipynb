```python
import pyecharts.options as opts
from pyecharts.charts import Line

"""
Gallery 使用 pyecharts 1.1.0
参考地址: https://echarts.apache.org/examples/editor.html?c=area-stack
目前无法实现的功能:
暂无
"""


x_data = ["Incubation Period", "Increase Period", "High Plateau Period", "Recession Period"]
y_data = [0.24, 0.35, 0.28, 0.29]


```

```python
stackline=(
        Line()
        .add_xaxis(xaxis_data=x_data)
        .add_yaxis(
            series_name="上海",
            stack="权重",
            y_axis=[0.24, 0.35, 0.28, 0.29],
            areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
            label_opts=opts.LabelOpts(is_show=False),
        )
        .add_yaxis(
            series_name="方舱",
            stack="权重",
            y_axis=[0.15, 0.18, 0.24,  0.17],
            areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
            label_opts=opts.LabelOpts(is_show=False),
        )
        .add_yaxis(
            series_name="隔离",
            stack="权重",
            y_axis=[ 0.06,  0.04, 0.04, 0.04],
            areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
            label_opts=opts.LabelOpts(is_show=False),
        )
        .add_yaxis(
            series_name="医院",
            stack="权重",
            y_axis=[0.10, 0.096, 0.10, 0.07],
            areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
            label_opts=opts.LabelOpts(is_show=False),
        )
        .add_yaxis(
            series_name="感染者",
            stack="权重",
            y_axis=[0.07, 0.03, 0.03, 0.07],
            areastyle_opts=opts.AreaStyleOpts(opacity=0.5),
            label_opts=opts.LabelOpts(is_show=True, position="top"),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Keywords change trend"),
            tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="cross"),
            yaxis_opts=opts.AxisOpts(
                type_="value",
                axistick_opts=opts.AxisTickOpts(is_show=True),
                splitline_opts=opts.SplitLineOpts(is_show=True),
            ),
            xaxis_opts=opts.AxisOpts(type_="category", boundary_gap=False),
        )
)
```

```python
stackline.render_notebook()
```

```python
stackline.render("分时数据\关键词堆积图.html")
```
