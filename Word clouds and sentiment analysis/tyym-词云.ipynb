```python
import pandas as pd
import jieba.analyse
from wordcloud import WordCloud,ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# 读取文件
pd_data = pd.read_csv('分时数据/Incubation Period.csv',low_memory=False)
```

```python
#读取内容
text = pd_data['微博正文'].tolist()#Pandas的tolist()函数用于将一个系列或数据帧中的列转换为列表。
```

```python
# 切割分词
wordlist = jieba.lcut_for_search(''.join(text)) 
result = ' '.join(wordlist)
```

```python
stopwords = [line.strip() for line in open('stopwords-master\stopwords.txt', 'r',encoding='utf-8').readlines()]
```

```python
wc = WordCloud(
    font_path='msyh.ttc',       # 中文
    background_color='white',    # 设置背景颜色为白色
    stopwords=stopwords,        # 设置禁用词，在生成的词云中不会出现set集合中的词
    #mask=im
    width=12000,
    height=8000,
    max_words=300
)
wc.generate(result)
plt.figure(figsize=(6, 4),dpi=300)
plt.imshow(wc)
plt.axis('off')
```

```python
#从词库中删除停用词库的内容
ciyun_words = ''
 
res = result.split()
preword=''
for word in res:
     if len(word)>1:#删去单字
            preword = preword + word+'\n'
            
preword2 = preword.split('\n')
for word in preword2:
    if word not in stopwords:
        ciyun_words += word #这是个列表
        
#for word in ciyun_words:
#    ciyunwords += 
        
ciyun_words = ' '.join(ciyun_words)#列表转换成字符串        
```

```python
# 设置参数，创建WordCloud对象
wc = WordCloud(
    font_path='msyh.ttc',       # 中文
    background_color='white',    # 设置背景颜色为白色
    stopwords=stopwords,        # 设置禁用词，在生成的词云中不会出现set集合中的词
    #mask=im
    width=12000,
    height=8000,
    max_words=300
)
wc.generate(ciyun_words)
plt.figure(figsize=(6, 4),dpi=300)
plt.imshow(wc)
plt.axis('off')
```
