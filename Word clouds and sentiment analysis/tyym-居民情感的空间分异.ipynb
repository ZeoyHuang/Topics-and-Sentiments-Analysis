```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Map
```

```python

# input data
df_input = pd.read_csv('分时数据\分区数据\情感平均值.csv',encoding='utf-8')
```

```python
districts = df_input['市区'].tolist()
value = df_input['情绪指数'].tolist()
max_value = max(value)
min_value = min(value)
```

```python
my_mark = "情感指数"
my_city = "上海"
my_title = "Sentiment index of all districts"
my_subtitle = "As of April 30, 2022"
```

```python
c = (
        Map(init_opts=opts.InitOpts(width ='1000px', height='500px'))
        .add(my_mark, [list(z) for z in zip(districts, value)], my_city)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False),is_map_symbol_show=False)
        .set_global_opts(
            title_opts=opts.TitleOpts(title=my_title, subtitle=my_subtitle),
            visualmap_opts=opts.VisualMapOpts(min_=min_value, 
                                              max_=max_value,
                                              range_text=["Positive", "Negative"],
                                              #is_piecewise=True,
                                              #range_color=["#565656",'#3a3a3a',"#ffc001"]
                                             ),
        )
    )
c.render_notebook()
#c.render('分时数据\分区数据\上海疫情情感地图1.html')
```

```python
from pyecharts.charts import Bar
```

```python
b = (
    Bar()
    .add_xaxis(districts)
    .add_yaxis("Sentiment index",value,color="#ffc001")#自定义颜色
    .set_global_opts(
        title_opts={"text": "Sentiment index of all districts"},
        brush_opts=opts.BrushOpts(), # 设置操作图表的画笔功能
        toolbox_opts=opts.ToolboxOpts(), # 设置操作图表的工具箱功能
        yaxis_opts=opts.AxisOpts(name="市区"), # 设置Y轴名称、定制化刻度单位
        xaxis_opts=opts.AxisOpts(name="Sentiment index"), # 设置X轴名称
       
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(is_show=False),
        markline_opts=opts.MarkLineOpts(
            data=[opts.MarkLineItem(x=0.5,)],
            linestyle_opts=opts.LineStyleOpts(width = 3,color = '#ef9a9a',type_='dashed') 
        ),
    )
    .reversal_axis()
    #.set_global_opts(title_opts=opts.TitleOpts(title="各区情绪指数"))
#     .render("标准柱状图.html")
)
b.render_notebook()
#b.render('分时数据\分区数据\各区情绪指数.html')
```
