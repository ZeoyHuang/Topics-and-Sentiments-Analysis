```python
import pyecharts.options as opts
from pyecharts.charts import WordCloud
from pyecharts.globals import SymbolType
import pandas as pd
import jieba
from jieba import analyse
```

```python
data=pd.read_csv('分时数据\分区数据\长宁区.csv',encoding = 'utf-8')
```

```python
stop_words = stopwords = [line.strip() for line in open('stopwords-master\stopwords.txt', 'r',encoding='utf-8').readlines()]
text = ''
for i in data['微博正文']:
    if i not in stop_words:
        text+=i
    
key_words = jieba.analyse.set_stop_words('stopwords-master\stopwords.txt')    #设置停用词
key_words = jieba.analyse.extract_tags(sentence=text, topK=50, withWeight=True, allowPOS=())
key_words
```

```python
worldcloud=(
     WordCloud()
     .add(series_name="权重", data_pair=key_words, word_size_range=[20, 100])
     .set_global_opts(
         title_opts=opts.TitleOpts(
             title="长宁区",title_textstyle_opts=opts.TextStyleOpts(font_size=23)
            ),
         tooltip_opts=opts.TooltipOpts(is_show=True),
         )
     )
worldcloud.render_notebook()
```
