```python
import pandas as pd
```

```python
data = pd.read_csv('分时数据/上海疫情（汇总版）.csv',encoding = 'utf-8')
data['发布时间'] = pd.to_datetime(data['发布时间'])#very important!
```

```python
data.head()
```

```python
open_day1='2022/3/1'
open_day2='2022/3/24'
open_day3='2022/4/10'
open_day4='2022/4/22'
open_day5='2022/5/1'
```

```python
data_incu = data[(data['发布时间']>=open_day1) & (data['发布时间']<open_day2)]
```

```python
data_incr = data[(data['发布时间']>=open_day2) & (data['发布时间']<open_day3)]
```

```python
data_high = data[(data['发布时间']>=open_day3) & (data['发布时间']<open_day4)]
```

```python
data_rece = data[(data['发布时间']>=open_day3) & (data['发布时间']<open_day5)]
```

```python
data_incu.to_csv('分时数据/Incubation Period.csv',encoding = 'utf-8_sig')
data_incr.to_csv('分时数据/Increase Period.csv',encoding = 'utf-8_sig')
data_high.to_csv('分时数据/High Plateau Period.csv',encoding = 'utf-8_sig')
data_rece.to_csv('分时数据/Recession Period.csv',encoding = 'utf-8_sig')
```
